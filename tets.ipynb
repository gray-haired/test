```python
import numpy as np
import pandas as pd
import re
```

```python
# загружаем данные
df = pd.read_csv("/content/sample_data/test_data.csv")
# оставляем только реплики менеджера
df_manager = df[df['role'] == 'manager']
df_manager
```

```python
# выберем реплики с приветствием
# попробуем учесть несколько вариантов приветствий
greeting = []

for item in df_manager['text']:
    if "здравствуйте" in item.lower():
        greeting.append(True)
    elif "добрый день" in item.lower():
        greeting.append(True)
    elif "добрый вечер" in item.lower():
        greeting.append(True)
    elif "доброе утро" in item.lower():
        greeting.append(True)
    elif "приветствую" in item.lower():
        greeting.append(True)
    else:
        greeting.append(False)

# создадим столбец и отметим реплики с приветствиями
df_manager['greeting'] = greeting

# оставим только интересующие нас столбцы и выведем все фразы, где менеджер
# поприветствовал клиента
greeting_manager = df_manager[['dlg_id','role','text','greeting']]
print('Реплики менеджера с приветствием\n')
greeting_manager = greeting_manager[greeting_manager.greeting == True]
greeting_manager
```

```python
# выберем реплики где менеджер представляется
# попробуем учесть несколько вариантов представлений

introduction = []

for item in df_manager['text']:
    if "меня зовут" in item.lower():
        introduction.append(True)
    elif "вас беспокоит" in item.lower():
        introduction.append(True)
    elif "это менеджер" in item.lower():
        introduction.append(True)
    else:
        introduction.append(False)

# по аналогии с приветственными репликами
df_manager['introduction'] = introduction
intro_manager = df_manager[['dlg_id','role','text','introduction']]
print('Реплики, где менеджер представился\n')
intro_manager = intro_manager[intro_manager.introduction == True]
intro_manager
```

```python
# получим имя менеджера.
# чаще всего менеджер представляется начиная со слов "меня зовут"
# найдем имя, используя регулярное выражение
# перед этим сделав сооветствующее преобразование реплик

manager_text = '. '.join(df_manager['text'].tolist())
match = re.findall(r"меня зовут \w*", manager_text, re.I)
manager_name = ' '.join(match)

# теперь разобьем строку и получим элемент, который мы считаем именем
name = re.split(r' ', manager_name, 3)
name[2]
```

```python
# наименование компании получим аналогичным образом
# допуская, что при указании компании менеджер начинает ее со слов "компания"
# или с указания ее юридической формы (АО, ЗАО, ООО, ИП)

company = []

for item in df_manager['text']:
    if "компания" in item.lower():
        company.append(True)
    elif "ооо" in item.lower():
        company.append(True)
    elif "зао" in item.lower():
        company.append(True)
    elif "ип" in item.lower():
        company.append(True)
    else:
        company.append(False)

df_manager['company'] = company
companys_name = df_manager[['dlg_id','role','text','company']]
print('Реплики, где менеджер назвал компанию\n')
companys_name = companys_name[companys_name.company == True]
companys_name
```

```python
# выделим реплики, где менеджер попрощался

goodbye = []

for item in df_manager['text']:
    if "до свидания" in item.lower():
        goodbye.append(True)
    elif "всего доброго" in item.lower():
        goodbye.append(True)
    elif "всего хорошего" in item.lower():
        goodbye.append(True)
    else:
        goodbye.append(False)

df_manager['goodbye'] = goodbye
bye_manager = df_manager[['dlg_id','role','text','goodbye']]
print('Реплики, где менеджер попрощался\n')
bye_manager = bye_manager[bye_manager.goodbye == True]
bye_manager
```
