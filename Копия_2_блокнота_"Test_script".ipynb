```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/sample_data/test_data.csv")
```

```python
data.head()
```

```python
data = data[data['role'] == 'manager']
data.head()
```

```python
insight = []
goodbye = []
introduction = []

for item in data['text']:
    if "здравствуйте" in item.lower():
        insight.append(True)
    else:
        insight.append(False)
        
for item in data['text']:
    if "меня зовут" in item.lower():
        introduction.append(True)
    else:
        introduction.append(False)
        
for item in data['text']:
    if "до свидания" in item.lower():
        goodbye.append(True)
    else:
        goodbye.append(False)

data['insight'] = insight
data['goodbye'] = goodbye
data['introduction'] = introduction
data.head()
```

```python
greeting = data[['dlg_id','role','text','insight']]
print('Реплики менеджера с приветствием\n')
greeting[greeting.insight == True]
```

```python
parting = data[['dlg_id','role','text','goodbye']]
print('Реплики, где менеджера попрощался\n') 
parting[parting.goodbye == True]
```

```python
intro = data[['dlg_id','role','text','introduction']]
print('Реплики, где менеджер представился\n') 
intro[intro.introduction == True]
```

```python
text_series = '. '.join(data['text'].tolist())
text_cap = text_series.title()
text_cap
```

```python
from natasha import (
    Segmenter,
    MorphVocab,
    
    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    NewsNERTagger,
    
    PER,
    LOC,
    NamesExtractor,
    DatesExtractor,
    MoneyExtractor,
    AddrExtractor,

    Doc
)

segmenter = Segmenter()
morph_vocab = MorphVocab()

emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
syntax_parser = NewsSyntaxParser(emb)
ner_tagger = NewsNERTagger(emb)

names_extractor = NamesExtractor(morph_vocab)
dates_extractor = DatesExtractor(morph_vocab)
money_extractor = MoneyExtractor(morph_vocab)
addr_extractor = AddrExtractor(morph_vocab)
```

```python
doc = Doc(text_cap)
doc.segment(segmenter) 
doc.tag_morph(morph_tagger)
doc.parse_syntax(syntax_parser)
doc.tag_ner(ner_tagger)
```
